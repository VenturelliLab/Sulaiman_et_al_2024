```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from scipy.stats import norm

import os
import time
import itertools

from autode.autode import ODE
```

# Define generalized Lotka Volterra (gLV) system of differential equations 

```python
# using gLV model 
def system(t, x, params, ctrl_params): 
    
    # determine system dimension
    dim = len(x)
    # growth rates
    r = np.reshape(params[:dim], x.shape)
    # interaction terms
    A = np.reshape(params[dim:], [dim,dim])

    # derivative of x 
    return x * (r + A@x) 
```

```python
# import file names
files = os.listdir("data/")
files = [f for f in files if "processed.csv" in f]
files
```

# fit gLV models

```python
exp_names = []
models = {}
for file in files:
    # import data
    df = pd.read_csv(f"data/{file}")

    # determine species names 
    species = df.columns.values[2:]

    # define initial parameter guess
    dim = len(species)
    n_params = dim + dim**2 
    params = np.zeros(n_params)

    # set growth rates positive
    params[:dim] = 0.3
    params[dim:] = 0.0
    k = dim 
    # set Aii terms negative to stabilize initial guess
    for i in range(dim):
        for j in range(dim):
            if i == j:
                params[k] = -1.0
            k += 1

    # instantiate gLV fit 
    model = ODE(system = system, 
                df=df, 
                sys_vars = species, 
                params = params) 

    # fit to data 
    t0 = time.time()
    model.fit()
    print("Elapsed time {:.2f}s".format(time.time()-t0))
    
    # add model to models dict
    models[file.split("_")[0]] = model
```

```python
sp_names = df.columns.values[2:]
sp_names
```

```python
# define design space to search for CD inhibiting communities
# create matrix of all possible communities
numspecies = len(sp_names) 
X = np.array([np.reshape(np.array(i), (1, numspecies)) for i in itertools.product([0, 1], repeat = numspecies)])
X = np.squeeze(X, 1)

# only keep combinations that have CD 
CD_inds = X[:, 4] > 0
X = X[CD_inds]

# normalize so that sum of OD is .01
X = (.01 * X.T / np.sum(X, 1)).T

# define evaluation times
t_eval = np.linspace(0, 24)
```

```python
# determine for each strain the AUC of CD in mono-culture
AUC = {}
for i,strain in enumerate(["DSM", "MS001", "MS008", "MS014"]):
    comm_pred, comm_stdv, _ = models[strain].predict(X[0], t_eval)
    cd_pred = comm_pred[:, 4]
    
    AUC[strain] = np.trapz(y=cd_pred, x=t_eval)
```

```python
# define objective to compute AUC of CD given a community
def objective(comm, t_eval, strains):
    # init objective value 
    obj = 0.
    
    # loop over set of strains 
    for strain in strains:
        comm_pred, comm_stdv, _ = models[strain].predict(comm, t_eval)
        cd_pred = comm_pred[:, 4]
        cd_stdv = comm_stdv[:, 4]
        
        # objective is the amount that CD was reduced compared to mono culture
        obj += (AUC[strain] - np.trapz(y=cd_pred, x=t_eval)) / len(strains)
        
    return obj
```

```python
def comm_name(present_species):
    c_name = ""
    for p_s in present_species:
        if p_s != 'CD':
            c_name += p_s + "-"
    return c_name[:-1]
```
