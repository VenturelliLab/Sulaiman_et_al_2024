```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from scipy.stats import norm

import os
import time
import itertools

from glove.model import *
```

```python
# import file names
files = os.listdir("data/")
strains = ["DSM", "MS001", "MS008", "MS014"]
```

# fit gLV models

```python
exp_names = []
for strain in strains:
    # import DTL 0 data
    df = pd.concat([pd.read_csv(f"data/{fname}") for fname in files if strains[0] in fname and "EXP0019" in fname])
    
    # determine species names 
    species = df.columns.values[2:]
    
    # list of parameter names 
    param_names = []
    for s1 in species:
        for s2 in species:
            param_names += [s1+"*"+s2]
    param_names = list(species) + param_names

    # instantiate gLV fit 
    model0 = gLV(species, df)

    # fit to data 
    model0.fit()
    
    # import designed data
    df_universal = pd.read_csv("data/Universal.csv")
    df_strain = pd.concat([pd.read_csv(f"data/{fname}") for fname in files if strain in fname])
    df = pd.concat((df_universal, df_strain))
    df.sort_values(by=["Treatments", "Time"], inplace=True)

    # arrange species in same columns as before 
    columns = ["Treatments", "Time"] + list(species)
    df = df[columns].copy()

    # instantiate gLV fit 
    model = gLV(species, df)

    # fit to data 
    model.fit()
    
    # plot parameter distribution
    n_species = len(species)
    Avec0 = model0.params[n_species:]
    Avec = model.params[n_species:]
    Aij_std0 = np.sqrt(np.diag(model0.Ainv))[n_species:]
    Aij_std = np.sqrt(np.diag(model.Ainv))[n_species:]

    plt.figure(figsize=(18,18))
    k = 0
    for i in range(n_species):
        for j in range(n_species):
            plt.subplot(n_species, n_species, k+1)
            left = np.min([Avec0[k]-np.std(Avec0), Avec[k]-np.std(Avec)])
            right = np.max([Avec0[k]+np.std(Avec0), Avec[k]+np.std(Avec)])
            a = np.linspace(left, right)
            plt.plot(a, norm.pdf(a,Avec0[k],Aij_std0[k]), linestyle='--')
            plt.plot(a, norm.pdf(a,Avec[k],Aij_std[k]))
            plt.axvline(x=0, c='k', alpha=.5)
            k += 1
            if j == 0:
                plt.ylabel(species[i], fontsize=18)
            if i == n_species-1:
                plt.xlabel(species[j], fontsize=18)

    plt.suptitle(strain, fontsize=24)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.savefig("figures/"+strain+".pdf")
    plt.show()
    plt.close()
```
